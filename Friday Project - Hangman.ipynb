```python
from random import randint
from IPython.display import clear_output

class Hangman():
    def __init__(self):
        self.categories = {'birds': ['penguin','flamingo','pigeon','parrot','ostrich'], 
                           'countries': ['canada','mexico','argentina','switzerland','mongolia']}
        self.lives = 7
        self.scaffold = [[' ',' ',' ',' ','_',' ',' ',' ',' ',' ',' ',' '],
                         [' ',' ',' ','|',' ','|',' ',' ',' ',' ',' ',' '],
                         [' ',' ',' ',' ',' ','|',' ',' ',' ',' ',' ',' '],
                         [' ',' ',' ',' ',' ','|',' ',' ',' ',' ',' ',' '],
                         [' ',' ',' ',' ',' ','|',' ',' ',' ',' ',' ',' '],
                         [' ',' ',' ',' ',' ','|',' ',' ',' ',' ',' ',' '],
                         [' ',' ',' ',' ',' ','|',' ',' ',' ',' ',' ',' '],
                         [' ',' ',' ',' ',' ','|',' ',' ',' ',' ',' ',' '],
                         ['-']*12]
        self.positions = [[2,3],[3,3],[3,2],[3,4],[4,3],[5,2],[5,4]]
        self.stick_figure = ['O','|','\\','/','|','/','\\']
    
    def showCategories(self):
        print('\nCategories:')
        for category in self.categories.keys():
            print(f'{category.title()}')
    
    def chooseWord(self, category):
        words = self.categories[category]
        word = words[randint(0,len(words)-1)]
        return word
    
    def guessWord(self, word):
        try:
            letter = self.chooseLetter()
            if letter not in word and self.guess.count(letter) < 2:
                self.lives -= 1
                self.changeScaffold()
        except:
            pass
        
        blanks = []
        for w in word:
            if w not in self.guess:
                blanks.append('_')
            else:
                blanks.append(w.upper())
        clear_output()
        return blanks
        
    def chooseLetter(self):
        letter = input('\nChoose a letter: ').strip().lower()
        if letter.isalpha() and len(letter) == 1:
            self.guess.append(letter)
            return letter
        else:
            print('Choose a single, valid letter')
    
    def showScaffold(self):
        print(f'\nLives: {self.lives}')
        for row in self.scaffold:
            print(''.join(row))
    
    def changeScaffold(self):
        index = 6 - self.lives 
        position = self.positions[index]
        self.scaffold[position[0]][position[1]] = self.stick_figure[index]
    
    def clearScaffold(self):
        for p in self.positions:
            self.scaffold[p[0]][p[1]] = ' '
        
    def startGame(self):
        play = input('Play hangman? (y/n) ').strip().lower()
        while play == 'y':
            self.clearScaffold()
            self.lives = 7
            self.showCategories()
            category = input('\nWhich category? ').strip().lower()
            if category not in self.categories:
                print('Choose a valid category')
            else:
                word = self.chooseWord(category)
                self.showScaffold()
                print(f'\nWord:\n{" ".join(["_"]*len(word))}')
                self.guess = []
                while self.lives > 0:
                    blanks = self.guessWord(word)
                    print(f'\nGuesses: {" ".join([g.upper() for g in list(set(self.guess))])}')
                    print(f'\nWord:\n{" ".join(blanks)}')
                    self.showScaffold()
                    if '_' not in blanks:
                        break
                if self.lives == 0:
                    play = input('\nOut of lives. New game? ').strip().lower()
                else:
                    play = input('\nWord complete. New game? ').strip().lower()
                clear_output()
        print('\nGame ended.')

game = Hangman()
game.startGame()
```
